**단어 + 띄어쓰기 맞추기** 

ex) **문체부(tab)문화체육관광부** (예시)로 구성된 클리닝 파일(클리닝.txt) 구성 후 업로드

```
# 원본 text 파일에 적용하는 것이기 때문에 이전단계에서 해도 상관없으며, 이미 처리된 상황에 대해서는 중복할 필요없음

replacelist = []

# 파일을 읽어서, 바꾸기리스트를 만든다.
with open('/content/khaiii/rsc/src/클리닝.txt', "r") as f:
  for lines in f:
    변경전, 변경후 = lines.split('\t')
    변경후 = 변경후[:-1]
    변경전 = 변경전.replace("\xa0", " ")
    변경후 = 변경후.replace("\xa0", " ")
    replacelist.append((변경전, 변경후))

# 리스트의 단어쌍을 불러서, 바꾸기를 실행한다.
for 변경전, 변경후 in replacelist:
  df["내용"] = df["내용"].str.replace(변경전, 변경후)
```

```
# 필요 모듈 불러오기
import numpy as np
import pandas as pd
import itertools as iter
```

```
# khaiii 함수 정의 : 명사 동사 형용사 추출
def parse(sentence):
    pos = ((morph.lex, morph.tag) for word in api.analyze(sentence) for morph in word.morphs if morph.tag in ['NNG', 'VV', 'VA', 'NNP'])    
    words = [item[0] if item[1] == 'NNG' or item[1] == 'NNP' 
             else f'{item[0]}다' for item in pos]
    return words
```

```
# 문단을 문장 단위로 나누기
split = df['내용'].str.split(".")
split = split.apply(lambda x: pd.Series(x))
split = split.stack().reset_index(level=1, drop= True).to_frame('sentences')
df=df.merge(split, left_index= True, right_index=True, how="left")
df = df.drop(['내용'], axis = 1)
df['sentences'].replace("", np.nan, inplace = True)
df['sentences'].replace(" ", np.nan, inplace = True)
df.dropna(subset=['sentences'], inplace = True)
df = df.reset_index(drop= True) #인덱스 초기화
df
```

```
# parse함수에서 정의한 형태소만 뽑기
                                  
df['nlp'] = df['sentences'].apply(parse)
df
```

**불용어 처리**

```
# 학습 pretraining
from gensim.models.word2vec import Word2Vec

model = Word2Vec(df['nlp'].values, sg=1, window=3, min_count=1, workers=4, iter=100) #sg=(0=Cbow/1=skip-gram), window, min_count=제외할 최소 빈도수, worker= cpu core, iter= 반복수
model.init_sims(replace=True) # 학습 완료시 필요없는 메모리 unload
model_name ="word2vec_이명박"
model.save(model_name)
```

**공출현 매트릭스 형성**
(안되는 이유를 모르겠음)

```
import collections
import pandas as pd
import numpy as np

def co_occurrence(sentences, window_size):
    d = collections.defaultdict(int)
    vocab = set()
    for text in sentences:
        # preprocessing (use tokenizer instead)
        text = text.lower().split()
        # iterate over sentences
        for i in range(len(text)):
            token = text[i]
            vocab.add(token)  # add to vocab
            next_token = text[i+1 : i+1+window_size]
            for t in next_token:
                key = tuple( sorted([t, token]) )
                d[key] += 1

    # formulate the dictionary into dataframe
    vocab = sorted(vocab) # sort vocab
    df = pd.DataFrame(data=np.zeros((len(vocab), len(vocab)), dtype=np.int16),
                      index=vocab,
                      columns=vocab)
    for key, value in d.items():
        df.at[key[0], key[1]] = value
        df.at[key[1], key[0]] = value
    return df
```
